```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

###1.1. Скачать следующие данные: kc_house_data и laptop_price

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

###1.2. Изучите стоимости недвижимости


```python
sns.histplot(df['price'])
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество домов');
```

```python
sns.boxplot(data=df.price)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('')
plt.ylabel('Стоимость, $ млн')
plt.xticks([0], ['']);
```

На гистограмме и графике типа "ящик с усами" видно, что 50% домов стоят в пределах 300-6,5 тыс. долларов (2 и 3 квантиль), медиана составляет примерно 450 тыс. Дома стоимостью более 1.1 млн долларов можно характеризовать как выбросы.


###1.3. Изучите распределение квадратуры жилой


```python
sns.histplot(df['sqft_living'])
plt.title('Квадратура жилой площади')
plt.xlabel('Площадь, квадратные футы')
plt.ylabel('Количество домов');
```

```python
sns.boxplot(data=df.sqft_living)
plt.title('Распределение жилой площади')
plt.xlabel('')
plt.ylabel('Площадь, квадратные футы')
plt.xticks([0], ['']);
```

На гистограмме и графике типа "ящик с усами" видно, что площадь 50% домов лежит в пределах 1500-2500 квадратных футов (2 и 3 квантиль), медиана составляет примерно 2000 квадратных футов. Дома площадью более 4200 квадратных футов можно характеризовать как выбросы.

###1.4. Изучите распределение года постройки

## Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
plt.xlabel('Жилая площадь')
plt.ylabel('Стоимость дома');
```

Видна положительная корреляция стоимости дома с жилой площадью.

```python
sns.scatterplot(x=df['sqft_above'], y=df['price'])
plt.title('Зависимость цены дома от площади чердака')
plt.xlabel('Площадь чердака')
plt.ylabel('Стоимость дома');
```

Наличие и площадь чердака так же положительно влияют на стоимость дома, однако корреляция выражена меньше (более острый угол потенциальной кривой тренда)

```python
sns.barplot(x=df.grade, y=df.price)
plt.xlabel('Оценка состояния дома')
plt.ylabel('Цена дома, млн $');
```

Чем выше оценка состояния дома, тем выше его стоимость. Однако разница не так сильно выражена при низких оценках (1-5). А самые высокие оценки (11-13) дают самый высокий рост цены.

```python
sns.boxplot(x=df['waterfront'], y=df['price'])
plt.xlabel('Вид на набережную')
plt.ylabel('Цена, млн $')
plt.title('Распределение стоимости дома в зависимости от наличия вида на набережную')
plt.xticks([0, 1], ['no', 'yes']);
```

У домов с наличием вида на набережную медианная цена выше. У домов без вида очень много выбросов - домов, стоимость которых так же высока, как и при наличии вида, т.е. определяется какими-то другими критериями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(15, 8))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5)

plt.title('Correlation matrix');
```

Самым наглядным графиком является тепловая карта для матрицы корреляций признаков датасета. Он показывает, что самый высокий коэффициент корреляции (0,7) между стоимостью дома, его жилой площадью и оценкой состояния. Немного ниже этот показатель для площади чердака (0,6) и количества ванных комнат (0,5). Самые низкий коэффициент корреляции (0,3) со стоимостью дома у количества спален, количества этажей, наличия вида на набережную, площади подвала и долготой. А вот ширина координат расположения дома никак не влияет на его стоимость.
